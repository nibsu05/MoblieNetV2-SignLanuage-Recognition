```python
import os
import cv2
import numpy as np

# Define the base directory for the dataset
base_dataset_dir = 'asl_alphabet_dataset/asl_alphabet_train/asl_alphabet_train'

# Initialize lists to store images and labels
images = []
labels = []

# Loop through each subdirectory (class label)
for class_name in os.listdir(base_dataset_dir):
    class_dir = os.path.join(base_dataset_dir, class_name)
    if os.path.isdir(class_dir):
        # Loop through each image file in the subdirectory
        for image_name in os.listdir(class_dir):
            if class_name in ['del','space']:
                continue
            image_path = os.path.join(class_dir, image_name)
            if image_path.lower().endswith(('.png', '.jpg', '.jpeg')):
                # Read the image
                img = cv2.imread(image_path)

                # Check if image was loaded successfully
                if img is None:
                    print(f"Warning: Could not load image {image_path}")
                    continue

                # Resize the image to 64x64 pixels
                img_resized = cv2.resize(img, (64, 64))

                # Convert to grayscale
                img_gray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)

                # Normalize pixel values to [0, 1]
                img_normalized = img_gray.astype(np.float32) / 255.0

                # Append to lists
                images.append(img_normalized)
                labels.append(class_name)

print(f"Loaded {len(images)} images and {len(labels)} labels.")

# Convert lists to NumPy arrays
X = np.array(images)
y = np.array(labels)

print(f"Shape of images array (X): {X.shape}")
print(f"Shape of labels array (y): {y.shape}")

# Display the unique labels found
unique_labels = np.unique(y)
print(f"Unique labels found: {unique_labels}")
print(f"Number of unique labels: {len(unique_labels)}")
```

```python
from skimage.feature import hog

# Initialize a list to store HOG features
hog_features = []

# Iterate through each image in the preprocessed image array X
for image in X:
    # Calculate HOG features for each image
    # Using parameters specified in the instructions
    features = hog(image, orientations=9, pixels_per_cell=(8, 8),
                   cells_per_block=(2, 2), transform_sqrt=True, block_norm='L2-Hys')
    # Append the computed HOG feature vector to the list
    hog_features.append(features)

# Convert the list of HOG features into a NumPy array
X_features = np.array(hog_features)

# Print the shape of the X_features array to verify the dimensions
print(f"Shape of extracted HOG features array (X_features): {X_features.shape}")
```

# Task
Implement a multi-class Support Vector Machine (SVM) using the One-vs-Rest strategy by defining an `OvRSVM` Python class. This class should manage multiple binary `LinearSVM` classifiers (which will also be implemented). Then, train this `OvRSVM` model using the balanced training data (`X_train_balanced`, `y_train_balanced`), make predictions on the test set (`X_test`), and evaluate the overall accuracy of the model, summarizing its performance and discussing future improvements.

## Implement One-vs-Rest Multi-class SVM

### Subtask:
Define a new Python class `OvRSVM` that orchestrates the training and prediction for multiple classes using the previously implemented `LinearSVM` as its base classifier. This class will manage `num_classes` instances of `LinearSVM`, training each to distinguish one specific class from all others.


**Reasoning**:
To implement the One-vs-Rest Multi-class SVM, I will define two classes: `LinearSVM` for binary classification as the base classifier and `OvRSVM` to orchestrate multiple `LinearSVM` instances for multi-class prediction, following the specified instructions for their methods and initialization.



```python
import numpy as np
from sklearn.metrics import accuracy_score

# --- CC HM H TR TON HC ---

# nh ngha hm Loss (Hinge Loss)
def hinge_loss(X, y, W, b, lambda_param):
    """Tnh ton Hinge Loss cho ton b tp d liu. y l hm chi ph (Cost Function) ca SVM."""
    n_samples = X.shape[0]

    # Tnh score (w.x - b): Gi tr u ra trc khi p dng hm sign.
    scores = np.dot(X, W) - b

    # Tnh margin (y * score): Khong cch ca mu n ng bin gii (margin > 0 l on ng).
    margin = y * scores

    # Hinge Loss: max(0, 1 - margin). Mt mt ch xy ra khi margin < 1.
    hinge_term = np.maximum(0, 1 - margin)

    # Hinge Loss + Regularization (Cost Function): Tng chi ph bao gm c vic gi trng s nh.
    # lambda_param * np.dot(W, W) l thut ng Regularization (L2 penalty).
    loss = np.mean(hinge_term) + lambda_param * np.dot(W, W)
    return loss

# nh ngha hm Accuracy
def calculate_accuracy(X, y, W, b):
    """Tnh ton  chnh xc (Accuracy) cho tp d liu."""
    # Tnh gi tr c tnh (Decision Function)
    approx = np.dot(X, W) - b

    # D on: S dng hm sign (du)  phn loi thnh 1 hoc -1.
    predictions = np.sign(approx)

    # So snh d on vi nhn thc t v tnh t l ng.
    return np.mean(predictions == y)

# --- CLASS: LINEARSVM (Cp nht  tnh ton 4 metrics Train/Val) ---
class LinearSVM:
    def __init__(self, learning_rate=0.001, lambda_param=0.01, n_iterations=1000):
        # Tham s tc  hc (Bc nhy khi cp nht)
        self.learning_rate = learning_rate
        # Tham s iu chnh (Regularization) - Ngn overfitting
        self.lambda_param = lambda_param
        # S ln lp qua ton b d liu (S epochs)
        self.n_iterations = n_iterations
        self.weights = None # Vector trng s W
        self.bias = None    # Bias term b
        # History: Lu 4 metrics Train/Val
        self.history_ = {'loss': [], 'accuracy': [], 'val_loss': [], 'val_accuracy': []}

    # Sa fit  nhn X_val, y_val
    def fit(self, X, y, X_val=None, y_val=None):
        n_samples, n_features = X.shape

        # CHUYN I NHN: m bo nhn l -1 hoc 1 cho tp train
        y = np.where(y <= 0, -1, 1)

        # m bo nhn l -1 hoc 1 cho tp validation (nu c)
        y_val_binary = np.where(y_val <= 0, -1, 1) if y_val is not None else None
        is_val_available = X_val is not None # C kim tra tp val

        self.weights = np.zeros(n_features)
        self.bias = 0

        # Gradient Descent (Ti u ha ngu nhin tng mu)
        for epoch in range(self.n_iterations):
            # Shuffle d liu sau mi epoch
            indices = np.arange(n_samples)
            np.random.shuffle(indices)
            X_shuffled = X[indices]
            y_shuffled = y[indices]

            for idx, x_i in enumerate(X_shuffled):
                y_i = y_shuffled[idx]

                # Kim tra margin
                condition = y_i * (np.dot(x_i, self.weights) - self.bias) >= 1

                if condition:
                    # TRNG HP 1: ng v ngoi Margin (Ch cp nht Regularization)
                    self.weights -= self.learning_rate * (2 * self.lambda_param * self.weights)
                else:
                    # TRNG HP 2: Sai hoc nm trong Margin (Cp nht Reg. + Loss)
                    self.weights -= self.learning_rate * (2 * self.lambda_param * self.weights - np.dot(x_i, y_i))
                    self.bias -= self.learning_rate * y_i

            # --- GHI LI HISTORY (4 METRICS) SAU MI EPOCH ---

            # 1. Train Metrics
            self.history_['loss'].append(hinge_loss(X, y, self.weights, self.bias, self.lambda_param))
            self.history_['accuracy'].append(calculate_accuracy(X, y, self.weights, self.bias))

            # 2. Validation Metrics
            if is_val_available:
                self.history_['val_loss'].append(hinge_loss(X_val, y_val_binary, self.weights, self.bias, self.lambda_param))
                self.history_['val_accuracy'].append(calculate_accuracy(X_val, y_val_binary, self.weights, self.bias))
            else:
                 # Nu khng c tp Val, thm None  gi kch thc mng
                 self.history_['val_loss'].append(None)
                 self.history_['val_accuracy'].append(None)


    def predict(self, X):
        # Tnh Decision Score (w.x - b)
        approx = np.dot(X, self.weights) - self.bias
        # Tr v kt qu phn loi nh phn (-1 hoc 1)
        return np.sign(approx)

# --- CLASS: OvRSVM (Cp nht  tng hp Validation History) ---
class OvRSVM:
    def __init__(self, learning_rate=0.001, lambda_param=0.01, n_iterations=100):
        self.learning_rate = learning_rate
        self.lambda_param = lambda_param
        self.n_iterations = n_iterations
        self.classifiers = {}
        self.classes = []
        self.history_ = {}

    # Cp nht fit  nhn c X_val, y_val
    def fit(self, X, y, X_val, y_val):
        self.classes = np.unique(y) # Xc nh cc lp duy nht
        n_epochs = self.n_iterations

        # Khi to history chung (Ma trn 2D: epochs x classes)
        self.history_ = {
            'loss': np.zeros((n_epochs, len(self.classes))),
            'accuracy': np.zeros((n_epochs, len(self.classes))),
            'val_loss': np.zeros((n_epochs, len(self.classes))),
            'val_accuracy': np.zeros((n_epochs, len(self.classes)))
        }

        for i, c in enumerate(self.classes):
            print(f"  -> Training classifier for class '{c}' ({i+1}/{len(self.classes)})...")

            # TO NHN NH PHN cho tp TRAIN v VAL
            y_binary_train = np.where(y == c, 1, -1)
            y_binary_val = np.where(y_val == c, 1, -1) # Nhn binary cho tp Val

            # Khi to v train LinearSVM con
            svm = LinearSVM(self.learning_rate, self.lambda_param, self.n_iterations)

            # Hun luyn, truyn c validation set vo LinearSVM
            svm.fit(X, y_binary_train, X_val=X_val, y_val=y_binary_val)

            # --- GHI LI HISTORY CON VO HISTORY CHUNG ---
            # Lu li 4 metrics t SVM con
            self.history_['loss'][:, i] = svm.history_['loss']
            self.history_['accuracy'][:, i] = svm.history_['accuracy']
            self.history_['val_loss'][:, i] = svm.history_['val_loss']
            self.history_['val_accuracy'][:, i] = svm.history_['val_accuracy']

            self.classifiers[c] = svm

        # Tnh ton gi tr trung bnh (MEAN) qua tt c cc b phn loi con
        self.history_['mean_loss'] = np.mean(self.history_['loss'], axis=1)
        self.history_['mean_accuracy'] = np.mean(self.history_['accuracy'], axis=1)
        self.history_['mean_val_loss'] = np.mean(self.history_['val_loss'], axis=1)
        self.history_['mean_val_accuracy'] = np.mean(self.history_['val_accuracy'], axis=1)

        print("OvRSVM training complete.")

    def predict(self, X):
        n_samples = X.shape[0]
        # Ma trn Scores: Lu tr im s quyt nh (confidence)
        scores = np.zeros((n_samples, len(self.classes)))

        # Ly im s t TT C cc b phn loi con
        for i, class_name in enumerate(self.classes):
            svm = self.classifiers[class_name]
            # im s l khong cch n ng bin gii: w.x - b
            scores[:, i] = np.dot(X, svm.weights) - svm.bias

        # TM KT QU CUI CNG: Chn lp c im s cao nht (Max Score)
        predicted_class_indices = np.argmax(scores, axis=1)

        # nh x ch mc v li tn nhn ban u (A, B, C...)
        predicted_labels = [self.classes[idx] for idx in predicted_class_indices]

        return np.array(predicted_labels)
print("LinearSVM and OvRSVM classes defined.")
```

```python
import numpy as np

# 1. Initialize an empty list to store the indices for the balanced dataset
balanced_indices = []

# 2. Get the unique class labels from y_train
unique_classes = np.unique(y_train)

# 3. Define the number of samples per class as 500
samples_per_class = 500

# 4. For each unique class label, find all indices and select samples
for class_name in unique_classes:
    # Find all indices in y_train that correspond to the current class
    class_indices = np.where(y_train == class_name)[0]

    # Check if there are enough samples in the class, if not, take all available samples
    num_samples_to_take = min(len(class_indices), samples_per_class)

    # From these indices, randomly select 'num_samples_to_take' indices
    # Use replace=False to ensure unique selections from the current class
    selected_indices = np.random.choice(class_indices, size=num_samples_to_take, replace=False)

    # Append the selected indices for the current class to the balanced_indices list
    balanced_indices.extend(selected_indices)

# 5. Concatenate all indices in balanced_indices and shuffle them to randomize the order
balanced_indices = np.array(balanced_indices)
np.random.shuffle(balanced_indices)

# 6. Use the balanced_indices to create X_train_balanced
X_train_balanced = X_train[balanced_indices]

# 7. Use the balanced_indices to create y_train_balanced
y_train_balanced = y_train[balanced_indices]

# 8. Print the shapes of X_train_balanced and y_train_balanced to confirm the balanced dataset size
print(f"Shape of X_train_balanced: {X_train_balanced.shape}")
print(f"Shape of y_train_balanced: {y_train_balanced.shape}")
print(f"Number of samples per class in balanced dataset: {samples_per_class} (or less if class had fewer)")
```

Now that the training data has been re-balanced to `X_train_balanced` and `y_train_balanced`, the next logical step is to initialize the `OvRSVM` model using default parameters and then train it with this re-balanced data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- HM H TR: TNH V IN PHN B LP ---
def get_class_distribution(y_data, name):
    """Tnh ton v nh dng phn b cc lp trong mt mng nhn."""
    # 1. m s ln xut hin ca mi nhn
    df = pd.Series(y_data).value_counts().reset_index()
    df.columns = ['Class Label', 'Number of Samples']

    # 2. Tnh phn trm
    total = len(y_data)
    df['Percentage (%)'] = (df['Number of Samples'] / total) * 100
    df['Percentage (%)'] = df['Percentage (%)'].round(2)

    # 3. Sp xp theo tn lp
    df = df.sort_values(by='Class Label').reset_index(drop=True)

    print(f"\n--- BNG THNG K CHI TIT: TP {name.upper()} ({total} mu) ---")
    print(df.to_markdown(index=False, floatfmt=".2f")) # Hin th bng chi tit
    print(f"Tng s lp: {len(df)}")
    return df

# --- HM CHNH: HIN TH TT C THNG K ---
def display_final_split_stats_detailed(X_train_bal, y_train_bal, X_val, y_val, X_test, y_test):
    """
    Thng k chi tit kch thc chung v phn b nhn ca tng tp (Train, Val, Test).
    """
    # --- 1. THNG K KCH THC CHUNG V T L SPLIT ---
    n_train_bal = len(X_train_bal)
    n_val = len(X_val)
    n_test = len(X_test)
    total = n_train_bal + n_val + n_test

    split_data = {
        'Tp d liu': ['Train ( cn bng)', 'Validation', 'Test', 'TNG CNG'],
        'S lng mu': [n_train_bal, n_val, n_test, total],
        'T l (%)': [
            n_train_bal / total * 100,
            n_val / total * 100,
            n_test / total * 100,
            100.0
        ]
    }
    df_split = pd.DataFrame(split_data)

    print("\n" + "="*80)
    print("THNG K T L CHIA TP D LIU CHUNG:")
    print("="*80)
    print(df_split.to_markdown(index=False, floatfmt=".2f"))
    print("="*80)

    # --- 2. THNG K S LNG MI NHN CHI TIT ---

    # A. Tp Train ( cn bng)
    df_train_bal = get_class_distribution(y_train_bal, "Train Balanced")

    # B. Tp Validation (Tp gc - Dng  kim tra tng qut)
    df_val = get_class_distribution(y_val, "Validation")

    # C. Tp Test (Tp gc)
    df_test = get_class_distribution(y_test, "Test")


    # --- 3. V BIU  (Tp Train  cn bng) ---

    plt.figure(figsize=(15, 6))

    sns.barplot(data=df_train_bal, x='Class Label', y='Number of Samples', palette='hls', hue='Class Label', legend=False)

    plt.title('Phn b s lng nh trong Tp Train ( Cn Bng)', fontsize=14)
    plt.xlabel('Nhn lp (Class Label)', fontsize=12)
    plt.ylabel('S lng mu', fontsize=12)
    plt.xticks(rotation=45)

    # V ng trung bnh/mc trn (500)
    mean_val = df_train_bal['Number of Samples'].mean()
    plt.axhline(mean_val, color='r', linestyle='--', label=f'Mc trung bnh: {mean_val:.0f}')
    plt.legend()

    plt.tight_layout()
    plt.show()

display_final_split_stats_detailed(X_train_balanced, y_train_balanced, X_val, y_val, X_test, y_test)
```

```python
import matplotlib.pyplot as plt

# --- 1. Sa hm Plotting  nhn d liu trung bnh t OvRSVM (C VAL) ---

def plot_training_history(history_data): # i tn tham s  trnh nhm ln
    """
    V biu  Accuracy v Loss da trn lch s trung bnh ca OvRSVM (Train v Validation).
    Yu cu: history_data phi cha 4 kha: 'mean_accuracy', 'mean_loss',
             'mean_val_accuracy', v 'mean_val_loss'.
    """
    # Ly d liu Train
    mean_acc = history_data['mean_accuracy']
    mean_loss = history_data['mean_loss']

    # Ly d liu Validation
    mean_val_acc = history_data['mean_val_accuracy']
    mean_val_loss = history_data['mean_val_loss']

    epochs = range(1, len(mean_acc) + 1)

    # To khung hnh vi 2 biu 
    plt.figure(figsize=(12, 5))

    # Biu  1: Accuracy ( chnh xc trung bnh)



    plt.subplot(1, 2, 1)
    # ng Train (Mu xanh)
    plt.plot(epochs, mean_acc, label='Train Accuracy', color='blue')
    # ng Validation (Mu , nt t)
    plt.plot(epochs, mean_val_acc, label='Validation Accuracy', color='red', linestyle='--')

    plt.title('Training vs. Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    # Biu  2: Loss (Hm mt mt trung bnh)
    plt.subplot(1, 2, 2)
    # ng Train
    plt.plot(epochs, mean_loss, label='Train Loss', color='blue')
    # ng Validation
    plt.plot(epochs, mean_val_loss, label='Validation Loss', color='red', linestyle='--')

    plt.title('Training vs. Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# --- LNH GI HM (Sau khi train xong) ---
# Kim tra s tn ti ca history_ trc khi gi
if hasattr(ovr_svm_model, 'history_'):
    plot_training_history(ovr_svm_model.history_)
else:
    print("Li: Thuc tnh 'history_' cha c to trong m hnh OvRSVM.")
```
